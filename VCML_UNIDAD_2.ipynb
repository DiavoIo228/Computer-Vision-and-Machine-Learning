```python
# Importar la clase Sequential
# Sequential permite crear redes neuronales capa por capa en orden secuencial
from tensorflow.keras.models import Sequential

# Importar la clase Dense
# Dense representa una capa totalmente conectada (fully connected layer)
from tensorflow.keras.layers import Dense


# Crear el modelo de red neuronal usando la estructura secuencial
# Esto significa que las capas se ejecutan una después de otra
Model = Sequential([

    # Primera capa (capa oculta)
    # Dense(16) → crea 16 neuronas en esta capa
    # activation='relu' → función de activación ReLU
    # input_shape=(10,) → la red recibe 10 valores de entrada
    #
    # Esto significa que cada muestra de entrada tiene 10 características
    Dense(16, activation='relu', input_shape=(10,)),


    # Segunda capa oculta
    # Dense(8) → crea 8 neuronas
    # activation='relu' → introduce no linealidad en el modelo
    #
    # Esta capa procesa la información generada por la capa anterior
    Dense(8, activation='relu'),


    # Capa de salida
    # Dense(1) → solo una neurona de salida
    # activation='sigmoid' → produce un valor entre 0 y 1
    #
    # Se utiliza principalmente para:
    # - clasificación binaria
    # - probabilidad de una clase
    Dense(1, activation='sigmoid')

])


# Mostrar el resumen del modelo
# summary() muestra:
# - número de capas
# - número de neuronas por capa
# - número de parámetros (pesos y bias)
# - forma de entrada y salida
Model.summary()
```

```python
# Importa la función para generar un conjunto de datos de clasificación sintético
from sklearn.datasets import make_classification

# Importa el modelo de red neuronal multicapa (Perceptrón Multicapa)
from sklearn.neural_network import MLPClassifier


# Genera un conjunto de datos artificial para clasificación
# n_samples=200 → crea 200 muestras (filas)
# n_features=4 → cada muestra tendrá 4 características (columnas)
# X → matriz de características (variables de entrada)
# y → vector de etiquetas (clases objetivo)
X, y = make_classification(n_samples=200, n_features=4)


# Crea el modelo de red neuronal
# hidden_layer_sizes=(10,10) → dos capas ocultas con 10 neuronas cada una
# max_iter=500 → máximo de 500 iteraciones durante el entrenamiento
clf = MLPClassifier(hidden_layer_sizes=(10,10), max_iter=500)


# Entrena el modelo usando los datos generados
# X → datos de entrada
# y → etiquetas reales
clf.fit(X, y)


# Evalúa el modelo usando los mismos datos de entrenamiento
# score() devuelve la exactitud (accuracy), es decir:
# (número de predicciones correctas / total de muestras)
print("precision:", clf.score(X, y))
```

```python
# Importa el algoritmo K-Means para clustering
from sklearn.cluster import KMeans

# Importa la librería NumPy para trabajar con arreglos y números aleatorios
import numpy as np


# Genera una matriz de datos aleatorios
# np.random.rand(100,2) crea:
# 100 filas (muestras)
# 2 columnas (características)
# Los valores están distribuidos uniformemente entre 0 y 1
X = np.random.rand(100, 2)


# Crea el modelo K-Means
# n_clusters=3 → queremos dividir los datos en 3 grupos (clusters)
kmeans = KMeans(n_clusters=3)


# Entrena el modelo con los datos
# El algoritmo:
# 1. Inicializa 3 centroides aleatorios
# 2. Asigna cada punto al centroide más cercano
# 3. Recalcula los centroides
# 4. Repite hasta converger
kmeans.fit(X)


# kmeans.labels_ contiene el cluster asignado a cada punto
# [:10] muestra solo las primeras 10 etiquetas
print(kmeans.labels_[:10])
```

```python
import numpy as np  # Se importa la librería NumPy (aunque en este código no se utiliza explícitamente)

def grad(x):
    """
    Calcula el gradiente de la función f(x) = x^2.

    La derivada de f(x) = x^2 es:
        f'(x) = 2x

    Parámetros:
        x (float): Valor actual de la variable.

    Retorna:
        float: El gradiente evaluado en x.
    """
    return 2 * x

# Valor inicial de x
x = 10

# Tasa de aprendizaje (learning rate)
lr = 0.1

# Bucle de descenso por gradiente
# Se actualiza x durante 10 iteraciones usando la regla:
# x = x - lr * grad(x)
for i in range(10):
    x = x - lr * grad(x)
    print(f"Iteracion {i+1}: x = {x}")
```

```python
import tensorflow as tf  # Se importa la librería TensorFlow

# Se crea una variable de TensorFlow con valor inicial 3.0
# tf.Variable permite que TensorFlow pueda calcular gradientes respecto a ella
x = tf.Variable(3.0)

# GradientTape registra las operaciones realizadas dentro de este bloque
# para poder calcular derivadas automáticamente (autodiferenciación)
with tf.GradientTape() as tape:
    y = x**2  # Definimos la función y = x^2

# Se calcula la derivada de y respecto a x
# Es decir, dy/dx
dy_dx = tape.gradient(y, x)

# Se imprime el valor numérico del gradiente
# .numpy() convierte el tensor en un valor de NumPy
print(dy_dx.numpy())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# Se importan dos métricas de evaluación desde sklearn:
# - accuracy_score: calcula la precisión del modelo
# - confusion_matrix: genera la matriz de confusión

# Valores reales (etiquetas verdaderas)
y_true = [0, 1, 1, 0, 1]

# Valores predichos por el modelo
y_pred = [0, 1, 0, 0, 1]

# Se calcula la precisión (accuracy)
# Accuracy = (predicciones correctas) / (total de muestras)
print("Accuracy:", accuracy_score(y_true, y_pred))

# Se calcula la matriz de confusión
# Formato en clasificación binaria:
# [[VN, FP],
#  [FN, VP]]
print("Matriz de confusion:\n", confusion_matrix(y_true, y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Se crea un modelo secuencial (las capas se agregan una tras otra)
model = Sequential([

    # Primera capa densa (fully connected)
    # 32 neuronas
    # Activación ReLU
    # input_shape=(10,) indica que el modelo recibe 10 características de entrada
    Dense(32, activation='relu', input_shape=(10,)),

    # Segunda capa oculta
    # 16 neuronas con activación ReLU
    Dense(16, activation='relu'),

    # Capa de salida
    # 1 neurona porque es clasificación binaria
    # Activación sigmoide → devuelve valores entre 0 y 1 (probabilidad)
    Dense(1, activation='sigmoid')
])

# Se configura el modelo para el entrenamiento
model.compile(
    optimizer='adam',                 # Optimizador Adam
    loss='binary_crossentropy',       # Función de pérdida para clasificación binaria
    metrics=['accuracy']              # Métrica de evaluación
)

# Muestra un resumen de la arquitectura del modelo
model.summary()
```

```python
# Importamos las librerías necesarias
# numpy: para operaciones matemáticas y manejo de arreglos numéricos
# matplotlib.pyplot: para generar gráficos
import numpy as np
import matplotlib.pyplot as plt

# Generamos 100 valores igualmente espaciados entre -5 y 5
# Esto servirá como dominio para las funciones de activación
x = np.linspace(-5, 5, 100)

# -------------------------
# Definición de funciones de activación
# -------------------------

# ReLU (Rectified Linear Unit)
# Devuelve 0 cuando x < 0
# Devuelve x cuando x >= 0
relu = np.maximum(0, x)

# Sigmoid
# Fórmula: 1 / (1 + e^(-x))
# Devuelve valores entre 0 y 1
sigmoid = 1/(1+np.exp(-x))

# Tangente hiperbólica (tanh)
# Devuelve valores entre -1 y 1
tanh = np.tanh(x)

# -------------------------
# Graficación
# -------------------------

# Graficamos cada función en el mismo plano
plt.plot(x, relu, label='ReLU')
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')

# Mostramos la leyenda para identificar cada curva
plt.legend()

# Agregamos una cuadrícula para facilitar la lectura
plt.grid(True)

# Mostramos la gráfica en pantalla
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Se crea un modelo secuencial (las capas se agregan una tras otra)
model = Sequential([

    # Primera capa densa (fully connected)
    # 32 neuronas
    # Activación ReLU
    # input_shape=(10,) indica que el modelo recibe 10 características de entrada
    Dense(32, activation='relu', input_shape=(10,)),

    # Segunda capa oculta
    # 16 neuronas con activación ReLU
    Dense(16, activation='relu'),

    # Capa de salida
    # 1 neurona porque es clasificación binaria
    # Activación sigmoide → devuelve valores entre 0 y 1 (probabilidad)
    Dense(1, activation='sigmoid')
])

# Se configura el modelo para el entrenamiento
model.compile(
    optimizer='adam',                 # Optimizador Adam
    loss='binary_crossentropy',       # Función de pérdida para clasificación binaria
    metrics=['accuracy']              # Métrica de evaluación
)

# Guardar el modelo entrenado en un archivo
# "modelo_mlp.keras" será el archivo donde se almacenará la arquitectura,
# los pesos y la configuración del modelo.
model.save("modelo_mlp.keras")

# Importamos la función load_model para poder cargar modelos guardados
from tensorflow.keras.models import load_model

# Cargamos el modelo desde el archivo guardado previamente
modelo_cargado = load_model("modelo_mlp.keras")

# Confirmamos que el modelo se cargó correctamente
print("Modelo cargado correctamente")
```

```python
# Importa la librería TensorFlow, utilizada para construir y entrenar modelos
# de aprendizaje automático y redes neuronales.
import tensorflow as tf

# Importa matplotlib para poder visualizar imágenes o gráficos.
import matplotlib.pyplot as plt


# Carga el dataset MNIST desde Keras.
# MNIST es un conjunto de datos clásico que contiene imágenes de dígitos escritos a mano (0-9).
# Devuelve dos conjuntos:
#   - Entrenamiento (x_train, y_train)
#   - Prueba (x_test, y_test)
#
# x = imágenes
# y = etiquetas (el número que representa cada imagen)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()


# Imprime la forma (shape) del conjunto de entrenamiento.
# x_train.shape devuelve una tupla con:
# (cantidad_de_imágenes, alto, ancho)
print("Train:", x_train.shape)

# Imprime la forma (shape) del conjunto de prueba.
print("Test:", x_test.shape)
```

```python
# Muestra la primera imagen del conjunto de entrenamiento
# x_train[0] accede a la primera imagen almacenada en el dataset
# cmap='gray' indica que la imagen se mostrará en escala de grises
plt.imshow(x_train[0], cmap='gray')

# Coloca un título encima de la imagen
# f-string permite insertar dinámicamente el valor de la etiqueta
# y_train[0] es la etiqueta correspondiente a la primera imagen
plt.title(f"Etiqueta: {y_train[0]}")

# Oculta los ejes (números y líneas del plano cartesiano)
# para que la imagen se vea limpia
plt.axis("off")

# Renderiza y muestra la figura en pantalla
plt.show()
```

```python
# Normaliza los valores de los píxeles de las imágenes de entrenamiento.
# Las imágenes normalmente tienen valores entre 0 y 255.
# Al dividir entre 255 los valores quedan entre 0 y 1, lo cual
# ayuda a que los modelos de machine learning entrenen mejor.
x_train = x_train / 255

# Normaliza también las imágenes del conjunto de prueba.
x_test = x_test / 255


# Reestructura (reshape) las imágenes del conjunto de entrenamiento.
# Cada imagen de 28x28 píxeles se convierte en un vector de 784 valores (28*28).
# -1 indica que Python calculará automáticamente el número de filas.
x_train = x_train.reshape(-1, 28 * 28)

# Reestructura las imágenes del conjunto de prueba de la misma forma.
# Cada imagen pasa de 28x28 a un vector de 784 características.
x_test = x_test.reshape(-1, 28 * 28)


# Imprime la forma (dimensiones) del conjunto de entrenamiento.
# Esto permite verificar que el reshape se aplicó correctamente.
print(x_train.shape)

# Imprime la forma del conjunto de prueba.
print(x_test.shape)
```

```python
# Importa el modelo secuencial de Keras.
# Sequential permite crear redes neuronales donde las capas
# se agregan una después de otra en orden.
from tensorflow.keras.models import Sequential

# Importa la capa Dense.
# Dense es una capa totalmente conectada (fully connected),
# donde cada neurona se conecta con todas las neuronas de la capa anterior.
from tensorflow.keras.layers import Dense


# Se crea el modelo secuencial con tres capas
model = Sequential([

    # Primera capa oculta con 128 neuronas.
    # activation='relu' introduce no linealidad para aprender patrones complejos.
    # input_shape=(784,) indica que cada entrada tiene 784 características
    # (por ejemplo, una imagen de 28x28 convertida en vector).
    Dense(128, activation='relu', input_shape=(784,)),

    # Segunda capa oculta con 64 neuronas y activación ReLU.
    Dense(64, activation="relu"),

    # Capa de salida con 10 neuronas.
    # activation='softmax' se usa para clasificación multiclase,
    # ya que devuelve probabilidades para cada una de las 10 clases.
    Dense(10, activation="softmax")
])


# Muestra un resumen de la arquitectura del modelo:
# - capas
# - número de parámetros
# - dimensiones de salida
model.summary()
```
